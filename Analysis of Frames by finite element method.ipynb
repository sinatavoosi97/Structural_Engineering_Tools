```python
def spacing():
    s=(20*" ")
    return s

E=int(input("please enter the modulus of elasticity (psi) :"))
print(spacing())

N=int(input("please enter the number of elements :"))
print(spacing())

A=[]
for i in range(1,N+1):
    area=int(input(f"please enter the area of element {i} (in^2) :"))
    A.append(area)
print(spacing())

I=[]
for i in range(1,N+1):
    Inertia=int(input(f"please enter the Inertia of  element {i} (in^4) :"))
    I.append(Inertia)
print(spacing())
    
 
L=[]
for i in range(1,N+1):
    length=int(input(f"please enter the Length of  element {i} (ft) :"))
    L.append(length)
print(spacing())

X=[]
for i in range(1,N+1):
    x1=int(input(f"please enter the x1 of  element {i} (ft) :"))
    x2=int(input(f"please enter the x2 of  element {i} (ft) :"))
    X.append(x1)
    X.append(x2)
print(spacing())

Y=[]
for i in range(1,N+1):
    y1=int(input(f"please enter the y1 of  element {i} (ft) :"))
    y2=int(input(f"please enter the y2 of  element {i} (ft) :"))
    Y.append(y1)
    Y.append(y2)
print(spacing())

```
